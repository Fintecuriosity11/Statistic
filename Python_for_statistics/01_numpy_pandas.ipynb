# 등차수열 (Arithmetic progression)

\begin{align}
a_n= a_1 + (n -1)d
\end{align}

a_n =	the nᵗʰ term in the sequence<br>
a_1 = the first term in the sequence<br>
d= the common difference between terms
<br><br>

등차수열이란 {1,2,3,4,5}와 같이 앞뒤 값의 차이가 동일한 수열이다. 아래의 예시의 경우 각 값간의 차이가 1인 등차수열이다. <br>
<br>
인수의 시작위치는 start, 종료위치는 stop, 차잇값은 step<br><br>
**stop**으로 지정한 값은 배열에 포함되지 않는다.<br><br>
1부터 5까지 숫자를 넣을 것이기 때문에 stop에는 5에 1을 더한 6을 지정

```
np.arange(start = 1, stop = 6, step = 1)    # 시작이 1이고, 차이가 1인 등차수열 {1,2,3,4,5}
```

```
np.arange(start = 0.1, stop = 1.0, step = 0.2) 
```

```
np.tile("A", 5)  #같은 값을 잔뜩 채운 배열은 np.tile 함수 사용, "A" 5개
```

```
np.tile("0",4)
```
