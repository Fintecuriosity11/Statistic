# 2 다변량 데이터 (Multivariate Date)

* 여러 개의 변수(1개 이상의 변수)를 조합한 데이터를 다변량 데이터라고 부른다.

* 먼저 다변량 데이터는 '깔끔한 데이터'라는 형식으로 정리하는 것이 중요.
> '**깔끔한 데이터(Tidy Data)**'는 분석하기 쉽게 정리한 표 형태의 데이터이다. 2014년 해들리 위컴이 통계 소프트웨어 저널(Journal of statistical)에서 처음 제시한 용어 (아래에 research gate 링크 참조*)

대체 텍스트

[**Tidy data** by Hadley Wickham](https://www.researchgate.net/publication/215990669_Tidy_data)

* 깔끔한 데이터(tidy data)는 4가지 특징을 가지고 있음


1.   개별 값이 하나의 셀을 이룬다
2.   개별 변수가 하나의 열을 이룬다
3. 개별 관측이 하나의 행을 이룬다
4. 개별 관측 유닛 유형이 하나의 표를 이룬다.

> 깔끔한 데이터는 '의미와 구조가 대응한다' 는 특징을 지님.

> 위의 1~3번 정의라면 '값<=>셀', '변수<=>열', '관측<=>행'으로 대응

> 깔끔한 데이터는 '복잡한 데이터'를 모을 때 통일성 있는 처리를 할 수 있다는 장점을 지님


```
import numpy as np
import pandas as pd
from pandas import DataFrame as df

sample_array = np.array([1,3,4,5,7])

df=pd.DataFrame({
  '물고기 종류' : ["A", "B", "A" , "B","A"],
  '물고기 몸길이' : sample_array * 2  
})

print(df)
```

# 2.1 교차분석표, 차분석표, 분할표 (contingency table)

* 1열은 가게 위치(프랜차이즈), 2열은 음식 메뉴, 3열은 음식이 팔린 수

```
sample_array = np.array([20,15,16,12])

df=pd.DataFrame({
  '가게 위치' : ["안암","안암" , "신촌", "신촌"],
  '음식 메뉴' : ["김치 볶음밥", "계란 볶음밥", "김치 볶음밥 " , "계란 볶음밥"],
  '음식이 팔린 수' : sample_array  
})

print(df)
```

* 지저분한 데이터는 '행이 변수의 의미를  갖는' 경향이 있음.

* 위의 데이터가 깔끔한 데이터가 아닌 이유는 행의 변수에  '가게 위치'라는 의미가 있기 때문임.

* 깔끔한 데이터는 '행 하나에 1개의 결과'가 있도록 정리




# 2.2 공분산 (Covariance)

* 공분산은 2개의 연속형 변수의 관계성을 확인하는 통계량.



1.   **공분산이 0보다 클 때**: 변수 한 쪽이 큰 값을 갖게 되면 다른 한쪽도 커짐
2.   **공분산이 0보다 작을 때**: 변수의 한 쪽이 큰 값을 갖게 되면 다른 한쪽은 작아짐
3. **공분산이 0일 때**: 변수 사이의 관계성이 없음


------

\begin{align}
\ cov(x,y) = \dfrac{1}{N} \sum_{i=1}^N (x_i-\mu )(y_i-\mu ) 
\end{align}

*여기서 μx와 μy는 각각 x와 y의 평균을 의미. N은 표본 크기, Cov는 Covariance(공분산)의 약자
